### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

print(loja1.head())
print(loja2.head())
print(loja3.head())
print(loja4.head())
```

#1. Análise do faturamento



```python

#Faturamento da primeira loja.#
faturamento_total1 = loja1['Preço'].sum()
print(f"Faturamento total da Loja 1: R${faturamento_total1:,.2f}")

#Faturamento da segunda loja.#
faturamento_total2 = loja2['Preço'].sum()
print(f"Faturamento total da loja 2: R${faturamento_total2:,.2f}")

#Faturamento da terceira loja.#
faturamento_total3 = loja3['Preço'].sum()
print(f"Faturamento total da loja 3: R${faturamento_total3:,.2f}")

#Faturamento da quarta loja.#
faturamento_total4 = loja4['Preço'].sum()
print(f"Faturamento total da loja 4: R${faturamento_total4:,.2f}")
```

```python
# importando gráficos para representação de faturamento por lojas.#
import matplotlib.pyplot as plt
Lojas = ('loja 1', 'loja 2', 'loja 3' ,'loja 4')
faturamento_geral = (faturamento_total1, faturamento_total2, faturamento_total3, faturamento_total4)
plt.bar(x= Lojas, height= faturamento_geral)
```

# 2. Vendas por Categoria

```python
# Vendas por categoria da loja 1
categorias1 = (loja1.groupby('Categoria do Produto')['Preço']
.agg(Valor_total_de_Vendas='sum', Quantidade_de_Vendas='count').reset_index())

print(categorias1)
```

```python
# Vendas por categoria da loja 2
categorias2 = (loja2.groupby('Categoria do Produto')['Preço']
.agg(Valor_total_de_Vendas='sum', Quantidade_de_Vendas='count').reset_index())

print(categorias2)
```

```python
# Vendas por categoria da loja 3
categorias3 = (loja1.groupby('Categoria do Produto')['Preço']
.agg(Valor_total_de_Vendas='sum', Quantidade_de_Vendas='count').reset_index())

print(categorias3)
```

```python
# Vendas por categoria da loja 4
categorias4 = (loja4.groupby('Categoria do Produto')['Preço']
.agg(Valor_total_de_Vendas='sum', Quantidade_de_Vendas='count').reset_index())

print(categorias4)
```

```python
# Gráfico de quntidade de vendas por categria. #
import matplotlib.pyplot as plt
import numpy as np

# Categorias iguais nas quatro lojas
categorias = categorias1['Categoria do Produto']
x = np.arange(len(categorias))  # posição no eixo X
largura = 0.2  # largura das barras

plt.figure(figsize=(12, 6))

plt.bar(x - 1.5 * largura, categorias1['Quantidade_de_Vendas'], width=largura, label='Loja 1')
plt.bar(x - 0.5 * largura, categorias2['Quantidade_de_Vendas'], width=largura, label='Loja 2')
plt.bar(x + 0.5 * largura, categorias3['Quantidade_de_Vendas'], width=largura, label='Loja 3')
plt.bar(x + 1.5 * largura, categorias4['Quantidade_de_Vendas'], width=largura, label='Loja 4')

plt.title('Quantidade de Vendas por Categoria - 4 Lojas')
plt.xlabel('Categoria do Produto')
plt.ylabel('Quantidade de Vendas')
plt.xticks(x, categorias, rotation=50)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Gráfico feito com ajuda de IA

import matplotlib.pyplot as plt

# Definindo uma paleta de cores consistente para todas as lojas
cores = ['#FF9999','#66B3FF','#99FF99','#FFCC99','#C2C2F0', '#FFB266', '#B3FFCC']

# Criando figura com 4 gráficos lado a lado
plt.figure(figsize=(18, 5))

# Lista com os dataframes e títulos
lojas = [
    (categorias1, "Loja 1"),
    (categorias2, "Loja 2"),
    (categorias3, "Loja 3"),
    (categorias4, "Loja 4")
]

# Criando cada gráfico
for i, (df,titulo) in enumerate(lojas, 1):
    plt.subplot(1, 4, i)  # 1 linha, 4 colunas
    plt.pie(
        df['Quantidade_de_Vendas'],
        labels=df['Categoria do Produto'],
        autopct='%1.1f%%',
        startangle=90,
        colors=cores[:len(df)],  # usa só as cores necessárias
        textprops={'fontsize': 9}
    )
    plt.title(titulo, fontsize=12)
    plt.axis('equal')

# Título geral da figura
plt.suptitle("Distribuição da Quantidade de Vendas por Categoria — 4 Lojas", fontsize=16, fontweight='bold')

plt.tight_layout(rect=[0, 0, 1, 0.95])  # ajusta o layout para não sobrepor o título
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# -----------------------------
# DADOS DAS 4 LOJAS
# -----------------------------
lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

faturamento_total = [
    loja1['Preço'].sum(),
    loja2['Preço'].sum(),
    loja3['Preço'].sum(),
    loja4['Preço'].sum()
]

media_avaliacao = [
    loja1['Avaliação da compra'].mean(),
    loja2['Avaliação da compra'].mean(),
    loja3['Avaliação da compra'].mean(),
    loja4['Avaliação da compra'].mean()
]

quantidade_vendas = [
    loja1.shape[0],
    loja2.shape[0],
    loja3.shape[0],
    loja4.shape[0]
]

frete_medio = [
    loja1['Frete'].mean(),
    loja2['Frete'].mean(),
    loja3['Frete'].mean(),
    loja4['Frete'].mean()
]

# -----------------------------
# DATAFRAME RESUMO
# -----------------------------
df_resumo = pd.DataFrame({
    'Loja': lojas,
    'Faturamento_Total': faturamento_total,
    'Media_Avaliacao': media_avaliacao,
    'Quantidade_Vendas': quantidade_vendas,
    'Frete_Medio': frete_medio
})

# -----------------------------
# FUNÇÕES AUXILIARES
# -----------------------------
def top_produtos(loja, top_n=3):
    contagem = loja['Produto'].value_counts()
    mais_vendidos = contagem.head(top_n)
    menos_vendidos = contagem.tail(top_n)
    return mais_vendidos, menos_vendidos

def top_categorias(loja, top_n=3):
    contagem = loja['Categoria do Produto'].value_counts()
    mais_vendidas = contagem.head(top_n)
    menos_vendidas = contagem.tail(top_n)
    return mais_vendidas, menos_vendidas

# -----------------------------
# RELATÓRIO EM MARKDOWN
# -----------------------------
report_md = f"""
# Relatório Completo de Análise das 4 Lojas

## Introdução
Este relatório analisa quatro lojas com o objetivo de avaliar desempenho financeiro, satisfação dos clientes,
logística e comportamento de vendas para apoiar decisões estratégicas.

## Resumo das Lojas
| Loja | Faturamento Total | Média Avaliação | Quantidade Vendas | Frete Médio |
|------|------------------|----------------|------------------|-------------|
| Loja 1 | R$ {df_resumo.loc[0,'Faturamento_Total']:,.2f} | {df_resumo.loc[0,'Media_Avaliacao']:.2f} | {df_resumo.loc[0,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[0,'Frete_Medio']:.2f} |
| Loja 2 | R$ {df_resumo.loc[1,'Faturamento_Total']:,.2f} | {df_resumo.loc[1,'Media_Avaliacao']:.2f} | {df_resumo.loc[1,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[1,'Frete_Medio']:.2f} |
| Loja 3 | R$ {df_resumo.loc[2,'Faturamento_Total']:,.2f} | {df_resumo.loc[2,'Media_Avaliacao']:.2f} | {df_resumo.loc[2,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[2,'Frete_Medio']:.2f} |
| Loja 4 | R$ {df_resumo.loc[3,'Faturamento_Total']:,.2f} | {df_resumo.loc[3,'Media_Avaliacao']:.2f} | {df_resumo.loc[3,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[3,'Frete_Medio']:.2f} |

## Produtos Mais e Menos Vendidos
"""

for i, loja in enumerate([loja1, loja2, loja3, loja4]):
    mais, menos = top_produtos(loja)
    report_md += f"### {lojas[i]}\n"
    report_md += f"**Mais Vendidos:**\n{mais.to_dict()}\n\n"
    report_md += f"**Menos Vendidos:**\n{menos.to_dict()}\n\n"

# Categorias mais e menos vendidas
report_md += "## Categorias Mais e Menos Vendidas\n"
for i, loja in enumerate([loja1, loja2, loja3, loja4]):
    mais, menos = top_categorias(loja)
    report_md += f"### {lojas[i]}\n"
    report_md += f"**Mais Vendidas:**\n{mais.to_dict()}\n\n"
    report_md += f"**Menos Vendidas:**\n{menos.to_dict()}\n\n"

display(Markdown(report_md))

# -----------------------------
# GRÁFICOS COMPARATIVOS
# -----------------------------
cores = ['skyblue', 'lightgreen', 'salmon', 'violet']

# 1. Faturamento
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Faturamento_Total'], color=cores)
plt.title('Faturamento Total por Loja')
plt.ylabel('R$')
for i, v in enumerate(df_resumo['Faturamento_Total']):
    plt.text(i, v + 20000, f"R$ {v:,.0f}", ha='center', fontweight='bold')
plt.show()

# 2. Média de Avaliação
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Media_Avaliacao'], color=cores)
plt.title('Média de Avaliação por Loja')
plt.ylabel('Avaliação (0-5)')
plt.ylim(0,5)
for i, v in enumerate(df_resumo['Media_Avaliacao']):
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center', fontweight='bold')
plt.show()

# 3. Quantidade de Vendas
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Quantidade_Vendas'], color=cores)
plt.title('Quantidade de Vendas por Loja')
plt.ylabel('Número de Vendas')
for i, v in enumerate(df_resumo['Quantidade_Vendas']):
    plt.text(i, v + 50, f"{v:,}", ha='center', fontweight='bold')
plt.show()

# 4. Frete Médio
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Frete_Medio'], color=cores)
plt.title('Frete Médio por Loja')
plt.ylabel('R$')
for i, v in enumerate(df_resumo['Frete_Medio']):
    plt.text(i, v + 1, f"R$ {v:.2f}", ha='center', fontweight='bold')
plt.show()

```
