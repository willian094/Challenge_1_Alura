### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1
```

#1. Análise do faturamento



```python

#Faturamento da primeira loja.#
faturamento_total1 = loja1['Preço'].sum()
print(f"Faturamento total da Loja 1: R${faturamento_total1:,.2f}")

#Faturamento da segunda loja.#
faturamento_total2 = loja2['Preço'].sum()
print(f"Faturamento total da loja 2: R${faturamento_total2:,.2f}")

#Faturamento da terceira loja.#
faturamento_total3 = loja3['Preço'].sum()
print(f"Faturamento total da loja 3: R${faturamento_total3:,.2f}")
```

```python
# importando gráficos para representação de faturamento por lojas.#
import matplotlib.pyplot as plt
Lojas = ('loja 1', 'loja 2', 'loja 3')
faturamento_geral = (faturamento_total1, faturamento_total2, faturamento_total3)
plt.bar(x= Lojas, height= faturamento_geral)
```

# 6. Relatório de lojas para analize de venda de propriedade

Loja 1: maior faturamento
vendeu mais em valor, mesmo que a avaliação média seja um pouco menor.

Loja 2: avaliação média intermediária
faturamento um pouco menor.

Loja 3: avaliação média mais alta
faturamento menor.

Interpretação para investimento:

Se você prioriza lucro imediato, Loja 1 é mais atraente (maior faturamento).

Se você prioriza satisfação do cliente e reputação, Loja 3 é a melhor (maior avaliação média).

Loja 2 é um equilíbrio, mas não se destaca em nenhum dos dois pontos.


# **Observações técnicas:**

Faturamento Total

A Loja 1 tem o maior faturamento (≈1,53 milhões), seguida pela Loja 2 (≈1,49 milhões) e Loja 3 (≈1,46 milhões).

Isso indica que, em termos de receita bruta, a Loja 1 está performando melhor.

Média de Avaliação

A Loja 3 tem a melhor média de avaliação (≈4,048), seguida pela Loja 2 (≈4,037) e depois Loja 1 (≈3,977).

Isso sugere que os clientes da Loja 3 estão mais satisfeitos com a experiência de compra, mesmo com faturamento menor.

Quantidade de Vendas

Todas as lojas têm a mesma quantidade de vendas (2.359), o que indica que a diferença no faturamento vem principalmente do ticket médio (preço médio por venda).

Loja 1 vende produtos mais caros ou mais lucrativos por venda, já que faturamento é maior com a mesma quantidade de vendas.

## **Conclusão técnica para investimento/venda:**

Se o objetivo é maximizar receita rapidamente:
A Loja 1 é a mais interessante, pois gera mais faturamento, mesmo com avaliação levemente menor. O ticket médio alto ajuda a aumentar o lucro por venda.

Se o objetivo é fidelização e satisfação do cliente:
A Loja 3 pode ser estratégica, já que clientes mais satisfeitos tendem a comprar novamente, possibilitando crescimento sustentável no longo prazo.

Equilíbrio:
Loja 2 fica no meio-termo: faturamento razoável e avaliação relativamente alta. Pode ser uma opção “menos arriscada” para testar novos produtos ou expansão.

# Insight estratégico:
Você poderia usar a Loja 1 como principal geradora de receita e investir na melhoria da experiência do cliente (aumentando a avaliação) para potencializar ainda mais os lucros.

```python
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

# Dados das lojas (exemplo de cálculo)
lojas = ['Loja 1', 'Loja 2', 'Loja 3']
faturamento_total = [
    loja1['Preço'].sum(),
    loja2['Preço'].sum(),
    loja3['Preço'].sum()
]
media_avaliacao = [
    loja1['Avaliação da compra'].mean(),
    loja2['Avaliação da compra'].mean(),
    loja3['Avaliação da compra'].mean()
]
quantidade_vendas = [
    loja1.shape[0],
    loja2.shape[0],
    loja3.shape[0]
]
frete_medio = [
    loja1['Frete'].mean(),
    loja2['Frete'].mean(),
    loja3['Frete'].mean()
]

# Criar DataFrame resumo
df_resumo = pd.DataFrame({
    'Loja': lojas,
    'Faturamento_Total': faturamento_total,
    'Media_Avaliacao': media_avaliacao,
    'Quantidade_Vendas': quantidade_vendas,
    'Frete_Medio': frete_medio
})

# Função para produtos mais e menos vendidos
def top_produtos(loja, top_n=3):
    contagem = loja['Produto'].value_counts()
    mais_vendidos = contagem.head(top_n)
    menos_vendidos = contagem.tail(top_n)
    return mais_vendidos, menos_vendidos

# Função para categorias mais e menos vendidas
def top_categorias(loja, top_n=3):
    contagem = loja['Categoria do Produto'].value_counts()
    mais_vendidas = contagem.head(top_n)
    menos_vendidas = contagem.tail(top_n)
    return mais_vendidas, menos_vendidas

# Criar relatório em Markdown
report_md = f"""
# Relatório Completo de Análise das Lojas

## Introdução
Este relatório analisa três lojas com o objetivo de determinar qual delas seria mais vantajosa para venda e investimento.
Foram analisados faturamento, média de avaliação, quantidade de vendas, frete médio, produtos e categorias mais/menos vendidos.

## Resumo das Lojas
| Loja | Faturamento Total | Média Avaliação | Quantidade Vendas | Frete Médio |
|------|-----------------|----------------|-----------------|-------------|
| Loja 1 | R$ {df_resumo.loc[0,'Faturamento_Total']:,.2f} | {df_resumo.loc[0,'Media_Avaliacao']:.2f} | {df_resumo.loc[0,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[0,'Frete_Medio']:.2f} |
| Loja 2 | R$ {df_resumo.loc[1,'Faturamento_Total']:,.2f} | {df_resumo.loc[1,'Media_Avaliacao']:.2f} | {df_resumo.loc[1,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[1,'Frete_Medio']:.2f} |
| Loja 3 | R$ {df_resumo.loc[2,'Faturamento_Total']:,.2f} | {df_resumo.loc[2,'Media_Avaliacao']:.2f} | {df_resumo.loc[2,'Quantidade_Vendas']:,} | R$ {df_resumo.loc[2,'Frete_Medio']:.2f} |

## Produtos Mais e Menos Vendidos

"""

# Adicionar produtos
for i, loja in enumerate([loja1, loja2, loja3]):
    mais, menos = top_produtos(loja)
    report_md += f"### {lojas[i]}\n"
    report_md += f"**Mais Vendidos:**\n{mais.to_dict()}\n\n"
    report_md += f"**Menos Vendidos:**\n{menos.to_dict()}\n\n"

# Categorias mais e menos vendidas
report_md += "## Categorias Mais e Menos Vendidas\n"
for i, loja in enumerate([loja1, loja2, loja3]):
    mais, menos = top_categorias(loja)
    report_md += f"### {lojas[i]}\n"
    report_md += f"**Mais Vendidas:**\n{mais.to_dict()}\n\n"
    report_md += f"**Menos Vendidas:**\n{menos.to_dict()}\n\n"

# Exibir Markdown
display(Markdown(report_md))

# -----------------------------
# Gráficos comparativos
# -----------------------------
cores = ['skyblue', 'lightgreen', 'salmon']

# 1. Faturamento Total
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Faturamento_Total'], color=cores)
plt.title('Faturamento Total por Loja')
plt.ylabel('R$')
for i, v in enumerate(df_resumo['Faturamento_Total']):
    plt.text(i, v + 20000, f"R$ {v:,.0f}", ha='center', fontweight='bold')
plt.show()

# 2. Média de Avaliação
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Media_Avaliacao'], color=cores)
plt.title('Média de Avaliação por Loja')
plt.ylabel('Avaliação (0-5)')
plt.ylim(0,5)
for i, v in enumerate(df_resumo['Media_Avaliacao']):
    plt.text(i, v + 0.05, f"{v:.2f}", ha='center', fontweight='bold')
plt.show()

# 3. Quantidade de Vendas
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Quantidade_Vendas'], color=cores)
plt.title('Quantidade de Vendas por Loja')
plt.ylabel('Número de Vendas')
for i, v in enumerate(df_resumo['Quantidade_Vendas']):
    plt.text(i, v + 50, f"{v:,}", ha='center', fontweight='bold')
plt.show()

# 4. Frete Médio
plt.figure(figsize=(8,5))
plt.bar(df_resumo['Loja'], df_resumo['Frete_Medio'], color=cores)
plt.title('Frete Médio por Loja')
plt.ylabel('R$')
for i, v in enumerate(df_resumo['Frete_Medio']):
    plt.text(i, v + 1, f"R$ {v:.2f}", ha='center', fontweight='bold')
plt.show()
```
